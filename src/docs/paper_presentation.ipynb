```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def create_detailed_flowchart():
    fig, ax = plt.subplots(figsize=(14, 12))
    ax.set_xlim(0, 12)
    ax.set_ylim(0, 20)
    ax.axis('off')

    def draw_box(text, x, y, width=3, height=1, boxstyle="round,pad=0.3", color="lightgray"):
        box = mpatches.FancyBboxPatch((x, y), width, height, boxstyle=boxstyle, ec="black", fc=color)
        ax.add_patch(box)
        ax.text(x + width / 2, y + height / 2, text, ha="center", va="center", fontsize=10)
    
    def draw_arrow(x1, y1, x2, y2):
        ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                    arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8))

    # Zero-Shot Inference Flowchart
    draw_box("Start", 4, 18, color="lightgreen")
    draw_box("Load API Key and Test Data", 4, 16.5)
    draw_box("For Each Row in Test Data", 4, 15, boxstyle="round,pad=0.3")
    draw_box("Construct Prompt", 4, 13.5)
    draw_box("Send Prompt to OpenAI GPT Model", 4, 12)
    draw_box("Capture Response", 4, 10.5)
    draw_box("Extract Generated Text", 4, 9)
    draw_box("Store Results", 4, 7.5)
    draw_box("Return Results", 4, 6, color="lightgreen")

    draw_arrow(5.5, 18, 5.5, 17.5)
    draw_arrow(5.5, 16.5, 5.5, 16)
    draw_arrow(5.5, 15, 5.5, 14.5)
    draw_arrow(5.5, 13.5, 5.5, 13)
    draw_arrow(5.5, 12, 5.5, 11.5)
    draw_arrow(5.5, 10.5, 5.5, 10)
    draw_arrow(5.5, 9, 5.5, 8.5)
    draw_arrow(5.5, 7.5, 5.5, 7)

    # One-Shot Inference Flowchart
    draw_box("Prepare Example", 8, 16.5, color="lightblue")
    draw_box("Fixed Example: Use the first row", 8, 15)
    draw_box("Random Example: Select a random row\n for each query", 8, 13.5)

    draw_arrow(8.5, 16.5, 8.5, 16)
    draw_arrow(8.5, 15, 8.5, 14.5)
    draw_arrow(8.5, 13.5, 8.5, 13)

    # Connecting both flowcharts
    draw_arrow(5.5, 16, 7.5, 16)
    draw_arrow(5.5, 14.5, 7.5, 14.5)
    draw_arrow(8.5, 13, 5.5, 13)

    # Adding more details
    draw_box("Prompt", 0.5, 16.5, width=3, color="lightyellow")
    ax.text(0.20, 16, 
            ("Generate the adapted ending to fill these three aspects:\n"
             "1. Minimal Intervention:\n Adjust the story's original ending with \n minimal changes required to align \n  it with the counterfactual event.\n"
             "2. Narrative Insight:\n Understand the story structure and \n  make changes essential for maintaining \n coherence  and thematic consistency.\n"
             "3. Counterfactual Adaptability:\n Adapt the story's course in response \n to the counterfactual event that \n diverges from the initial event.\n"),
             ha="left", va="top", fontsize=8, wrap=True)

    draw_arrow(2.25, 15, 4, 15)

    plt.title("Detailed Data Flow for Zero-Shot and One-Shot Inference Using gpt-3.5-turbo-0125", fontsize=12)
    plt.show()

create_detailed_flowchart()
```
